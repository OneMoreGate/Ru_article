```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
import matplotlib.ticker as ticker
```

```python
energy_O = pd.read_csv('oxygen.csv', delimiter=';')
energy_Si = pd.read_csv('silicon.csv', delimiter=';')
energy_Ru = pd.read_csv('ruthenium.csv', delimiter=';')
```

```python
for j in energy_O.columns:
    energy_O[j] = [float(i.replace(',', '.')) for i in energy_O[j]] 

for j in energy_Si.columns:
    energy_Si[j] = [float(i.replace(',', '.')) for i in energy_Si[j]] 

for j in energy_Ru.columns:
    energy_Ru[j] = [float(i.replace(',', '.')) for i in energy_Ru[j]] 
```

```python
fig, ax = plt.subplots(figsize = [7,5])

ax.plot(energy_O['Binding Energy (E)'], energy_O['exp.'], 'k', linewidth = 1, label = 'XPS measurements')
ax.plot(energy_O['Binding Energy (E)'], energy_O['O1s A'], 'c--', linewidth = 1, label = 'Fiting')
ax.plot(energy_O['Binding Energy (E)'], energy_O['O1s B'], 'g--', linewidth = 1)
ax.plot(energy_O['Binding Energy (E)'], energy_O['Backgnd.'], 'y--', linewidth = 1)
ax.plot(energy_O['Binding Energy (E)'], energy_O['Envelope'], 'b--', linewidth = 1)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.2))
ax.yaxis.set_major_locator(ticker.MultipleLocator(100))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(20))
ax.set_xlim([527,536])
ax.set_ylim([900,1400])
ax.set_xlabel('Bonding energy, eV')
ax.get_yaxis().set_visible(False)
ax.invert_xaxis()
ax.legend()

#plt.savefig('Si2p_XPS.png', dpi = 300, bbox_inches = 'tight')
```

```python
fig, ax = plt.subplots(ncols=2, figsize = [10,5])

for i in range(len(ax)):
    ax[i].xaxis.set_major_locator(ticker.MultipleLocator(2))
    ax[i].xaxis.set_minor_locator(ticker.MultipleLocator(1))
    ax[i].get_yaxis().set_visible(False)
    ax[i].set_xlabel(r'Binding energy, eV')
    ax[i].spines['top'].set_visible(False)
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['left'].set_visible(False)

ax[0].plot(energy_Ru['Binding Energy (E)'], energy_Ru['Ru3d5'], 'b--', linewidth = 1.5)
ax[0].plot(energy_Ru['Binding Energy (E)'], energy_Ru['Ru3d3'], 'b--', linewidth = 1.5)
ax[0].plot(energy_Ru['Binding Energy (E)'], energy_Ru['Backgnd.'], 'g', linewidth = 1.5)
ax[0].plot(energy_Ru['Binding Energy (E)'], energy_Ru['Envelope'], 'k', linewidth = 1.5)
ax[0].plot(energy_Ru['Binding Energy (E)'], energy_Ru['exp.'], 'r', linewidth = 1.5)
ax[0].set_xlim([277,289])
ax[0].invert_xaxis()

ax[1].plot(energy_Si['Binding Energy (E)'], energy_Si['Si2p'], 'b--', linewidth = 1.5)
ax[1].plot(energy_Si['Binding Energy (E)'], energy_Si['Backgnd.'], 'g', linewidth = 1.5)
ax[1].plot(energy_Si['Binding Energy (E)'], energy_Si['Envelope'], 'k', linewidth = 1.5)
ax[1].plot(energy_Si['Binding Energy (E)'], energy_Si['exp.'], 'r', linewidth = 1.5, label = 'XPS measurements')
ax[1].set_xlim([99,107])
ax[1].invert_xaxis()

plt.savefig('XPS.png', dpi = 300, bbox_inches = 'tight')
```
